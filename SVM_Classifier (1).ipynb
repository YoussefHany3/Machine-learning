```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import csv
import numpy as np

# Load the iris dataset
df = pd.read_csv('/content/car_price_prediction.csv')
df
```

```python
X = df.drop(['Price','ID'] , axis = 1)
X
```

```python
X['Manufacturer'].unique()
X['Model'].unique()
X['Category'].unique()
X['Leather interior'].unique()
X['Fuel type'].unique()
X['Gear box type'].unique()
X['Drive wheels'].unique()
X['Doors'].unique()
X['Wheel'].unique()
X['Color'].unique()
X['Engine volume'].unique()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

X['Manufacturer']= label_encoder.fit_transform(X['Manufacturer'])
X['Model']= label_encoder.fit_transform(X['Model'])
X['Category']= label_encoder.fit_transform(X['Category'])
X['Leather interior']= label_encoder.fit_transform(X['Leather interior'])
X['Fuel type']= label_encoder.fit_transform(X['Fuel type'])
X['Gear box type']= label_encoder.fit_transform(X['Gear box type'])
X['Drive wheels']= label_encoder.fit_transform(X['Drive wheels'])
X['Doors']= label_encoder.fit_transform(X['Doors'])
X['Wheel']= label_encoder.fit_transform(X['Wheel'])
X['Color']= label_encoder.fit_transform(X['Color'])
X['Engine volume']= label_encoder.fit_transform(X['Engine volume'])



y = np.array(df['Price'])
X
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the SVM classifier with specific parameters
svm = SVC(C=1.0, kernel='rbf', degree=3, gamma='scale', coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None)

# Train the SVM classifier on the training set
svm.fit(X_train, y_train)

# Predict the labels of the testing set using the trained SVM classifier
y_pred = svm.predict(X_test)

# Calculate the accuracy score of the SVM classifier on the testing set
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```
